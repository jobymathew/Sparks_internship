# Task 3: Prediction using Decision Tree Algorithm

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import tree, datasets
```

```python
# Getting the dataset
iris_data = datasets.load_iris()
iris_df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
y = iris_data.target

print('The first five entries of the data\n', iris_df.head())
print('The class values \n',y)
```

## Decision Tree Training and visualisation

```python
# Preparing the data
X = iris_df.to_numpy()

# Defining the Decison Tree Classifier
dt_model = tree.DecisionTreeClassifier(random_state=0)
dt_model.fit(X, y)

fig = plt.figure(figsize=(27,20), facecolor='k')
dtree = tree.plot_tree(dt_model, 
                       feature_names=iris_data.feature_names, 
                       class_names=iris_data.target_names, 
                       filled=True, 
                       rounded=True,
                       fontsize=23)

# Customising the arrow in the tree
for object in dtree:
    arrow = object.arrow_patch
    if arrow is not None:
        arrow.set_edgecolor('yellow')
        arrow.set_linewidth(5)

```
